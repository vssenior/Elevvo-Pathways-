```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
sns.histplot(df['Age'], kde=True)
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df.isnull().sum()
```

```python
sns.histplot(df['Fare'], kde=True)
```

```python
df['Fare'] = df['Fare'].fillna(df['Fare'].mean())
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Fare'] = df['Fare'].astype(int)
df['Age'] = df['Age'].astype(int)
```

```python
sns.histplot(df['Ticket'], kde=True)
```

```python
df['Cabin'].unique()
```

```python
df['Cabin'] = df['Cabin'].fillna(df['Cabin'].mode()[0])   
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
                                    ###SUMMARY STATISTICS###


print(df.describe())
 
print(df.describe(include="object"))
```

```python
                                ###GROUPING BY GENDER###


print(df.groupby("Sex")["Survived"].mean())
print(df.groupby("Sex")["Survived"].value_counts())
```

```python
                            ###GROUPING BY PASSENGER CLASS###
# Survival rate by gender and class
print(df.groupby(["Sex", "Pclass"])["Survived"].mean())

# Survival counts by gender and class
print(df.groupby(["Sex", "Pclass"])["Survived"].value_counts())
```

```python


sns.barplot(data=df, x="Sex", y="Survived", hue="Pclass")
plt.title("Survival Rate by Gender & Class")
plt.show()
```

```python

sns.countplot(data=df, x="Sex", hue="Survived")
plt.title("Survival Counts by Gender")
plt.show()
```

```python
sns.barplot(data=df, x="Pclass", y="Survived")
plt.title("Average Survival Rate by Passenger Class")
plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.histplot(data=df, x="Age", hue="Survived", bins=30, kde=True)
plt.title("Age Distribution: Survivors vs Non-Survivors")
plt.show()
```

```python
sns.pairplot(
    df[["Age", "Fare", "Survived", "Pclass"]],
    hue="Survived"
)
plt.show()
```
